```python
# download the file
urllib.request.urlretrieve("https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user", "data/user.psv")
```

### Step 3

```python
users = pd.read_csv("data/user.psv", sep="|")
```

```python
users.head()
```

### Step 4

```python
# get mean age per occupation
users.groupby(['occupation'],as_index=False).mean()[['occupation','age']]
```

### Step 5

```python
# get male ratio per occupation
(users[users.gender == 'M'].groupby(['occupation']).agg(['count'])['user_id'] / users.groupby(['occupation']).agg(['count'])['user_id']).rename(columns={'count': 'male ratio'}).sort_values('male ratio',ascending=False)
```

### Step 6

```python
# get minimum ages of each occupation
users.groupby(['occupation']).min()[['age']].rename(columns={'age': 'min age'})
```

```python
# get maximum ages of each occupation
users.groupby(['occupation']).max()[['age']].rename(columns={'age': 'max age'})
```

### Step 7

```python
# get mean age of each combination of occupation and sex
(users.groupby(['occupation','gender']).mean()[['age']]).rename(columns={'age': 'mean age'})
```

### Step 8

```python
# get each occupation present the percentage of women and men
(users.groupby(['occupation','gender']).agg(['count'])['user_id'] / users.groupby(['occupation']).agg(['count'])['user_id'] * 100).rename(columns={'count': 'percentage of count'})
```

# Question 7

```python
# import dataset
us_mar_div = pd.read_csv("data/us-marriages-divorces-1867-2014.csv") 
```

```python
import seaborn as sns

marriages_per_capita = [us_mar_div.loc[us_mar_div['Year'] == 1900]["Marriages_per_1000"].values[0],
                       us_mar_div.loc[us_mar_div['Year'] == 1950]["Marriages_per_1000"].values[0],
                       us_mar_div.loc[us_mar_div['Year'] == 2000]["Marriages_per_1000"].values[0]]

divorces_per_capita = [us_mar_div.loc[us_mar_div['Year'] == 1900]["Divorces_per_1000"].values[0],
                       us_mar_div.loc[us_mar_div['Year'] == 1950]["Divorces_per_1000"].values[0],
                       us_mar_div.loc[us_mar_div['Year'] == 2000]["Divorces_per_1000"].values[0]]

sns_df = pd.DataFrame({
    'Year': ['1900', '1950', '2000'],
    'Marriages_per_1000': marriages_per_capita,
    'Divorces_per_1000': divorces_per_capita
})

fig, ax1 = plt.subplots(figsize=(10, 10))
tidy = sns_df.melt(id_vars='Year').rename(columns=str.title)
ax = sns.barplot(x='Year', y='Value', hue='Variable', data=tidy, ax=ax1)
ax.set(xlabel='Year', ylabel='Value per 1000 capita')
sns.despine(fig)
```
